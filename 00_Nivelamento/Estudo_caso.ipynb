```python
df_cleaned = df.dropna()
df_cleaned
```

```python
# Substituir vÃ­rgulas por pontos e converter a coluna 'mpg' para valores numÃ©ricos
df['mpg'] = df['mpg'].str.replace(',', '.').astype(float)

# Encontrar a linha com o valor mais alto na coluna 'mpg'
linha_mpg_maximo = df[df['mpg'] == df['mpg'].max()]

print(linha_mpg_maximo)
```

```python
df_cleaned['horsepower'] = df_cleaned['horsepower'].str.replace(',', '.').astype(float)

q3_horsepower = df_cleaned['horsepower'].quantile(0.75)

print(q3_horsepower)
```
